### Goal

## LTspice netlist and Python
**(file "controlled_Source.cir")**

```
* Controlled Source and Req - SPICE

VT 1 0 1
R1 1 0 1k
R2 1 2 500

* Fake source for ix
Vix 2 3 0
R3 3 0 1k

* CCCS
F1 0 3 Vix 2

* Circuit analysis
.OP

*saving the data
.PRINT V(1) I(VT)

.END
```


````python
import os
import ltspice
import matplotlib.pyplot as plt
import numpy as np


import warnings
warnings.filterwarnings("ignore")


#Run the simulation and save in binary .raw
os.system('/Applications/LTspice.app/Contents/MacOS/LTspice -b controlled_source.cir')
````

````python
#read the raw file generated by LTspice

filepath = 'controlled_source.raw'


l = ltspice.Ltspice(filepath)
l.parse() # Data loading sequence. It may take few minutes for huge file.

print(l.variables)
````

````python
var = l.variables

V_T = l.getData(var[0])
I_T = l.getData(var[1])

Req = -V_T/I_T

print('Req=',Req)
````
